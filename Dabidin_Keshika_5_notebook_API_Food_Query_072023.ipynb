```python
import requests
import pandas as pd

url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"

querystring = {"ingr":"champagne","category[0]":"grocery"}

headers = {
	"X-RapidAPI-Key": "36b0cadbc4msh8f35cdaf7f1ac47p1a274ejsna15f1e55dbab",
	"X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)
data = response.json()


print(data)
```

```python
import pprint
pp = pprint.PrettyPrinter(depth=4)
```

```python
data_copy=data
#del data_copy["text"]
#del data_copy["parsed"]
del data_copy['_links']
pp.pprint(data_copy)
```

```python
# Flatten the records first
def flatten_dict(d: dict) -> dict:
    out = {}

    def flatten(x, name: str = ''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x
    flatten(d)

    return out


data_flat = [flatten_dict(dat) for dat in data_copy['hints']]
df=pd.DataFrame(data_flat)
df.head()
```

```python
#Keeping the required columns 
df= df[['food_foodId', 'food_label','food_foodContentsLabel','food_image']]
df.head()
```

```python
# Renaming the columns
df=df.rename({'food_foodId': 'foodId', 'food_label': 'label','food_foodContentsLabel': 'foodContentsLabel','food_image': 'image'}, axis='columns')
df.head()
```

```python
url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"
categories = ['generic-foods','packaged-foods','Generic meals','fast-foods','grocery']
headers = {
	"X-RapidAPI-Key": "36b0cadbc4msh8f35cdaf7f1ac47p1a274ejsna15f1e55dbab",
	"X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
}
# Pour chaque catégorie, récupérer les 10 premiers produits et les stocker dans un fichier CSV
for category in categories:
    querystring = {"ingr":"","category[1]":category}#,"limit":"10"}
    response = requests.request("GET", url, headers=headers, params=querystring)
    data = response.json()
```

```python
data_copy=data
#del data_copy["_links"]
del data_copy["parsed"]
del data_copy["text"]
pp.pprint(data_copy)
```
