```python
train_data=pd.read_csv("fichier_train_with_labels.csv")
train_label=train_data.copy()
#train_label = train_label.set_index(['id'])
train_label.head()
```

```python
#Directories 
#labels = pd.read_csv('fichier_train_with_labels.csv')
labels=train_label

images_dir =r'C:\Users\kherm\Documents\Openclassrooms\projet 6\Flipkart\Images'
train = r"C:\Users\kherm\Documents\Openclassrooms\projet 6\Flipkart\train"
```

```python
if not os.path.exists(train):
    os.mkdir(train)

for filename, class_name in labels.values:
    # Create subdirectory with `class_name`
    if not os.path.exists(train + str(class_name)):
        os.mkdir(train + str(class_name))
    src_path = images_dir + '/'+ filename 
    dst_path = train + str(class_name) + '/' + filename
    try:
        shutil.copy(src_path, dst_path)
        print("sucessful")
    except IOError as e:
        print('Unable to copy file {} to {}'
              .format(src_path, dst_path))
    except:
        print('When try copy file {} to {}, unexpected error: {}'
              .format(src_path, dst_path, sys.exc_info()))
```
